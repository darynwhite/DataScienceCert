# Random numbers

**Often used in simulations.**

*<span style="color: brown;">In the example below, let's use simulation to find the numeric value of Ï€ (pi).</span>*

Estimates the value of pi using the Monte Carlo method.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import random
import matplotlib.pyplot as plt

def estimate_pi(total_points):
    inside_circle = 0
    x_points = []
    y_points = []
    colors = []

    for _ in range(total_points):
        x = random.random()
        y = random.random()

        if x**2 + y**2 <= 1:
            inside_circle += 1
            colors.append('lightgreen')  # Points inside the circle are green
        else:
            colors.append('red')  # Points outside the circle are red

        x_points.append(x)
        y_points.append(y)

    pi_estimate = 4 * inside_circle / total_points
    return pi_estimate, x_points, y_points, colors
```

```python
# Run the simulation with a large number of iterations for better accuracy
num_iterations = 5000
pi_estimate, x_points, y_points, colors = estimate_pi(num_iterations)

# Plot the unit square and the points
plt.figure(figsize=(6, 6))
plt.scatter(x_points, y_points, c=colors, s=1)
plt.plot(np.linspace(0, 1, 100), np.sqrt(1 - np.linspace(0, 1, 100)**2), color='black')
plt.plot([0, 1], [0, 0], color='black')
plt.plot([0, 0], [0, 1], color='black')
plt.title("Monte Carlo Estimation of Pi")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.show()

print("Estimated value of pi:", pi_estimate)
```

```python
# simplified (removed the graphics)

def estimate_pi(num_iterations):
    inside_circle = 0
    total_points = 0

    for _ in range(num_iterations):
        x = random.random()
        y = random.random()

        if x**2 + y**2 <= 1:
            inside_circle += 1

        total_points += 1

    pi_estimate = 4 * inside_circle / total_points
    return pi_estimate
```

```python
%%timeit -r 2 -n 8
# Run the simulation with a large number of iterations for better accuracy
num_iterations = 1000000
pi_estimate = estimate_pi(num_iterations)

print("Estimated value of pi:", pi_estimate)
```

```python
# Define the number of iterations for each simulation
num_iterations = 10000

# Run the simulation 1000 times
pi_estimates = [estimate_pi(num_iterations) for _ in range(1000)]

# Plot the histogram with KDE
sns.histplot(pi_estimates, bins=100, kde=True)
plt.title("Histogram of Pi Estimates with KDE")
plt.xlabel("Pi Estimate")
plt.ylabel("Density")
plt.grid()
plt.show()
```

```python
def estimate_pi_np(total_points):

    inside_circle = np.sum((np.random.rand(total_points))**2+(np.random.rand(total_points))**2 < 1) # {d^2 = x^2+y^2} < 1

    pi_estimate = 4 * inside_circle / total_points
    return pi_estimate
```

```python
%%timeit -r 2 -n 8
# Run the simulation with a large number of iterations for better accuracy
num_iterations = 1000000
pi_estimate = estimate_pi_np(num_iterations)

print("Estimated value of pi:", pi_estimate)
```
